```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score 
```

**Import Iris Dataset**

```python
iris = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
```

```python
iris.head()
```

```python
iris_ct = pd.crosstab(index = iris['species'],columns = 'count')
```

```python
x = iris.iloc[:,[0,1,2,3]].values
```

**Distribution plots**

```python
sns.boxenplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
sns.violinplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue = 'species')
plt.show()
```

```python
distortions = []
K = range(1,10)

for k in K:
    kmeansmodel = KMeans(n_clusters = k)
    kmeansmodel.fit(x)
    distortions.append(kmeansmodel.inertia_)
    
```

```python
plt.figure(figsize = (16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method Showing the Optimal CLusters')
plt.show()
```

We can see that the optimal number of cluster is : 3 

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
x[y_kmeans == 2, 0]
```

```python
#Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 20, c = 'purple', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 20, c = 'orange', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 20, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 70, c = 'red', label = 'Centroids')

plt.legend()
plt.show()
```

> x[y_kmeans == 0, 0]: This selects the x-coordinates of the data points that belong to cluster 0. It filters the x array based on the condition y_kmeans == 0 to only include the data points assigned to cluster 0, and then selects the first column (x-coordinate).

```python
x = iris.iloc[:,[0,1,2,3]].values
```

```python
y = iris.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# KMeans 
import random
random.seed(116)

kmeans = KMeans(n_clusters = 3)
y_pred = kmeans.fit_predict(x_test)

from sklearn.metrics import accuracy_score 
accuracy = accuracy_score(y_test,y_pred)

print('Accuracy for KMeans Clustering: ', accuracy)
```

```python
from sklearn.linear_model import LogisticRegression 
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
print('Accuracy for logistic regression: ', accuracy_score(y_test,y_pred))
```
